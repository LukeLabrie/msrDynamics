```python
from msrDynamics import Node, System
import numpy as np

# instantiate system 
f = System()

# instantiate nodes
x = Node(y0 = 1.0)
y = Node(y0 = 0.5)

x.dydt = y.y()
y.dydt = (1-x.y()*x.y())*y.y()-x.y()

# add nodes to system
f.add_nodes([x,y])
```

```python
T = np.arange(0,10,0.01)
sol = f.solve(T)
```
